```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# Load and prepare data
df = pd.read_csv("/training_data_source/csv/learn.csv")
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)

# Get the formula components
slope = reg.coef_[0]
intercept = reg.intercept_

# Create the visualization
plt.figure(figsize=(10, 6))
plt.scatter(df.area, df.price, color='red', marker='+', label='Actual Data Points')

# Generate line of best fit
x_range = np.linspace(df.area.min(), df.area.max(), 100)
y_range = slope * x_range + intercept
plt.plot(x_range, y_range, color='blue', label='Regression Line')

plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.title('House Prices: Actual vs Predicted')
plt.legend()

# Print the formula
print(f"Price = {slope:.2f} × Area + {intercept:.2f}")

prediction = reg.predict([[5000]])
print(f"\nPrediction for 5000 sq ft: ${prediction[0]:,.2f}")

r2_score = reg.score(df[['area']], df.price)
print(f"Model Accuracy (R²): {r2_score:.4f}")
```
