```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import os
```

```python

current_dir = os.getcwd()

base_dir = os.path.abspath(os.path.join(current_dir, '..', 'src',  'csv'))
file_path = os.path.join(base_dir, 'canada_per_capita_income.csv')

print("file path",file_path)

df = pd.read_csv(file_path)
df

df.head(10)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
prediction = reg.predict([[2027]])
```

```python
print(prediction)
```

```python
prediction = reg.predict([[2014]])
print(prediction)
```

```python
print(prediction)
```

```python
print(f"Slope (m): {reg.coef_[0]}")
print(f"Intercept (b): {reg.intercept_}")
```
