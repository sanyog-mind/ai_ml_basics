```python
import pandas as pd
```

```python
import os

current_dir = os.getcwd()

base_dir = os.path.abspath(os.path.join(current_dir, '..',  'csv'))
file_path = os.path.join(base_dir, 'carprices.csv')

print("file path",file_path)

df = pd.read_csv(file_path)
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
df_dummies = pd.concat([df, dummies], axis='columns')
df_dummies
```

```python
df_dummies.drop(['Car Model','Mercedez Benz C class'],axis=1,inplace=True)
df_dummies
```

```python
x = df_dummies.drop('Sell Price($)',axis='columns')
x
```

```python
y = df_dummies['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[45000,4,0,0]])
```

```python
# predict price of mercedes benz that is 4 years old with 45000 mileage 45000
reg.predict([[45000,5,0,0]])
```

```python
# predict price of BMW X5 that is 7 years old with mileage 86000
reg.predict([[86000,7,0,1]])
```

```python
reg.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = df
```

```python
# Converts the categorical column Car Model into numerical labels using LabelEncoder.
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
# Reason:
#
#     Machine learning models require numerical input. Encoding converts categories to numbers.
#     However, LabelEncoder assumes an ordinal relationship between the encoded values, which may not always be appropriate for categories without a natural order.
```

```python
dfle
```

```python
x = dfle[['Car Model','Mileage','Age(yrs)']].values
x
```

```python
y = dfle['Sell Price($)']
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ohe = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [0])
    ],
    remainder='passthrough'  # Keep other columns as they are
)

# Why is this Important for Machine Learning?
#
# Machine learning models, especially linear regression models like the one you're using, interpret the numerical values of features directly. If you don't use One-Hot Encoding, the model might:
#
#     Treat higher values as more "important" due to the numerical relationship.
#     Misinterpret the meaning of encoded numbers, leading to poor predictions.
#
# One-Hot Encoding removes this ambiguity by:
#
#     Assigning a distinct binary variable to each category.
#     Preventing the model from incorrectly assuming relationships between categories.
```

```python
x = ohe.fit_transform(x)
x
```

```python
reg.fit(x,y)
```

```python
# predict price of mercedes benz that is 4 years old with 45000 mileage 45000
reg.predict([[0,0,1,45000,4]])
```

```python
r2_score = reg.score(x, y)
print("R^2 Score:", r2_score)
```
