```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import os
```

```python

base_dir = os.path.abspath(os.path.join('..', 'csv'))

file_path = os.path.join(base_dir, 'homeprices.csv')
df = pd.read_csv(file_path)
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
df_dummies = pd.concat([df, dummies], axis='columns')
df_dummies
```

```python
df_dummies.drop(['town','west windsor'],axis=1,inplace=True)
df_dummies
```

```python
x = df_dummies.drop('price',axis='columns')
x
```

```python
y = df_dummies.price
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
# with 3400 sq.ft area in west windsor predict price of a home
reg.predict([[3400,0,0]])
```

```python
# with 2800 sq.ft area in robinsville predict price of a home
reg.predict([[2800,0,1]])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dfle = df
```

```python
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[['town','area']].values
x
```

```python
y = dfle.price.values
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ohe = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [0])  # Apply OneHotEncoder to the first column ('town')
    ],
    remainder='passthrough'  # Keep other columns as they are
)
```

```python
x = ohe.fit_transform(x)
x
```

```python
reg.fit(x,y)
```

```python
reg.predict([[0,0,1,3400]])
```

```python
r2_score = reg.score(x, y)
print("R^2 Score:", r2_score)
```
