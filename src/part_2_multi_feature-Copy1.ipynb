```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from word2number import w2n
import os
```

```python
current_dir = os.getcwd()

base_dir = os.path.abspath(os.path.join(current_dir, '..', 'src',  'csv'))
file_path = os.path.join(base_dir, 'hiring.csv')

print("file path",file_path)

df = pd.read_csv(file_path)
df
```

```python
import math
median_score = math.floor(df["test_score(out of 10)"].median())
median_score
```

```python
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(median_score)
df
```

```python
df['experience'] = df['experience'].replace(0, np.nan)
df
```

```python
df['experience'] = df['experience'].replace(np.nan,"zero")
df
```

```python
df['experience'] = df['experience'].apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
print("Prediction for [1, 3, 9]:",reg.predict([[1,3,9]]))
print("Prediction for [2, 3, 3]:", reg.predict([[2, 3, 3]]))
print("Prediction for [5, 9, 9]:", reg.predict([[5, 9, 9]]))
print("Prediction for [2, 6, 7]:", reg.predict([[2, 6, 7]]))
print("Prediction for [3, 8, 4]:", reg.predict([[3, 8, 4]]))
```
