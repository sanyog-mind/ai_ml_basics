```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    rate = 0.01  # Learning rate
    n = len(x)
    iterations = 10000
    y_history = []  # To store predictions for visualization

    plt.scatter(x, y, color='red', marker='+', linewidth=5)

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        if i % 500 == 0:  # Store every 500th iteration for visualization
            y_history.append((m_curr, b_curr, y_predicted))

        # Gradient calculation
        md = -(2/n) * sum(x * (y - y_predicted))
        yd = -(2/n) * sum(y - y_predicted)

        # Update parameters
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd

    # Plot the intermediate predictions
    for m, b, y_pred in y_history:
        plt.plot(x, y_pred, color='green', alpha=0.3)

    # Final prediction
    plt.plot(x, m_curr * x + b_curr, color='blue', label='Final Fit')
    plt.legend()
    plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
